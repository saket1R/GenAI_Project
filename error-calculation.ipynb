```python
import os
import pandas as pd
import re
import zipfile

# Define a function to extract the first numeric range from a string
def extract_average_range(value):
    # Find numeric ranges in the format of "number-number" (e.g., "10-20")
    ranges = re.findall(r'\b(\d+)-(\d+)\b', str(value))
    if ranges:
        # Calculate the average of the first range found
        start, end = map(int, ranges[0])
        return (start + end) / 2
    # If no range is found, return 0 as a default value
    return 0

# Folder containing the original CSV files
input_folder = '/kaggle/input/actual-pred'

# Output folder to store processed CSV files
output_folder = '/kaggle/working/error_cal'
os.makedirs(output_folder, exist_ok=True)

# Iterate through each file in the input folder
for file_name in os.listdir(input_folder):
    if file_name.endswith('.csv'):  # Process only CSV files
        input_file_path = os.path.join(input_folder, file_name)
        output_file_path = os.path.join(output_folder, file_name)
        
        # Read the CSV file
        df = pd.read_csv(input_file_path)
        
        # Ensure the required columns are present
        if 'prediction' in df.columns and 'correct_pred' in df.columns:
            # Calculate the average range for the "prediction" column
            df['prediction_numeric'] = df['prediction'].apply(extract_average_range)
            
            # Calculate the absolute error
            df['error'] = abs(df['prediction_numeric'] - df['correct_pred'])
            
            # Drop the temporary numeric column
            df.drop(columns=['prediction_numeric'], inplace=True)
            
            # Save the updated DataFrame to the output folder
            df.to_csv(output_file_path, index=False)
            print(f"Processed and saved: {file_name}")
        else:
            print(f"Required columns not found in {file_name}")

# Zip the output folder
zip_file_path = output_folder + '.zip'
with zipfile.ZipFile(zip_file_path, 'w', zipfile.ZIP_DEFLATED) as zipf:
    for root, _, files in os.walk(output_folder):
        for file in files:
            file_path = os.path.join(root, file)
            arcname = os.path.relpath(file_path, output_folder)
            zipf.write(file_path, arcname)

print(f"All files processed and zipped successfully at {zip_file_path}!")
```
